# Data Collection and Visualization

```python
# import libraries
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns


import hvplot.pandas
```

```python
#Read and display breast cancer data
BCW_df= pd.read_csv("breast-cancer.csv")
BCW_df.head(10)
```

```python
# run a frequrncy count on "diagnosis" column
BCW_df['diagnosis'].value_counts()
```

```python
#drop undesired columns for BCW_df
BCW_df=BCW_df.drop(columns=["id"])
```

```python
BCW_df.info()
```

```python
BCW_df.describe()
```

# Data PreProcessing

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
# encode the "diagnosis" column using a custom function
def encode_diagnosis(diagnosis):
    if diagnosis == "M":
        return 1
    else:
        return 0

#apply the encoded function to dataframe
BCW_df["diagnosis"]= BCW_df["diagnosis"]. apply(encode_diagnosis)

#Read BCW_Df
BCW_df.head()
```

### create X and y features

```python
#set the target of Y as "diagnosis"
y= BCW_df["diagnosis"]

#set the target of x by droping "tumour stage" column
X= BCW_df.drop(columns=["diagnosis"])

#Review X
X
```

```python
# Split data into training and testing set 
X_train, X_test, y_train, y_test = train_test_split(X, y,train_size= 0.8, random_state=1)
```

# Model Selection- SVM

```python
# Instantiate a linear SVM model
from sklearn.svm import SVC
classifier = SVC(kernel='linear')
classifier
```

# Model Training - SVM

```python
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# Fit the data
classifier.fit(X_train, y_train)
```

# Evaluation - SVM

```python
y_predicted=classifier.predict(X_test)
```

```python
accuracy_score(y_test,y_predicted)
```

# Model Selection- ANN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.filterwarnings("ignore", category=ConvergenceWarning)
```

```python
ann = Sequential()

#Hidden layer
ann.add(

    Dense( activation='relu', input_dim=30, units=10)
)
    
# Output layer
ann.add(Dense(units=1, activation='sigmoid'))
       
```

```python
#Compile the model
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
ann.summary()
```

# Model Training - ANN

```python
model_1 = ann.fit(X_train, y_train, validation_split=0.2, epochs=200)
```

# Evaluation - ANN

```python
train_acc = model_1.history['accuracy']
train_loss = model_1.history['loss']
valid_acc = model_1.history['val_accuracy']
valid_loss = model_1.history['val_loss']

epochs = [i+1 for i in range(len(train_acc))]

plt.figure(figsize=(20, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs, train_loss, 'r', label='Train Loss')
plt.plot(epochs, valid_loss, 'g', label='Valid Loss')
plt.title('Loss')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Loss')

plt.subplot(1, 2, 2)
plt.plot(epochs, train_acc, 'r', label='Train Accuracy')
plt.plot(epochs, valid_acc, 'g', label='Valid Accuracy')
plt.title('Accuracy')
plt.legend()
plt.xlabel('Epochs')
plt.ylabel('Accuracy')

plt.tight_layout()
plt.show()
```

```python
# predict the test data
y_pred = ann.predict(X_test)
y_pred = (y_pred > 0.5)
```

# Predictions/Results 

```python
# view confusion matrix
confusion_matrix = confusion_matrix(y_test, y_pred)

# Plot the confusion_matix
plt.figure(figsize=(6, 4))
sns.heatmap(confusion_matrix, annot=True, cmap="Blues", fmt="d")
plt.xlabel('Predicted labels')
plt.ylabel('True labels')
plt.title('Confusion Matrix')
plt.show()
```
